```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler, OneHotEncoder

# 1. Check and Apply Transformations for Skewness
# Visualize distributions to check skewness
plt.figure(figsize=(12, 8))

# Example for 'raisedhands'
plt.subplot(2, 2, 1)
sns.histplot(dataset['raisedhands'], bins=30, kde=True)
plt.title('Distribution of Raised Hands')

# Apply log transformation if skewed
dataset['raisedhands_Log'] = np.log1p(dataset['raisedhands'])

plt.subplot(2, 2, 2)
sns.histplot(dataset['raisedhands_Log'], bins=30, kde=True)
plt.title('Log Transformed Raised Hands')

# Example for 'VisITedResources'
plt.subplot(2, 2, 3)
sns.histplot(dataset['VisITedResources'], bins=30, kde=True)
plt.title('Distribution of Visited Resources')

# Apply square root transformation if skewed
dataset['VisITedResources_Sqrt'] = np.sqrt(dataset['VisITedResources'])

plt.subplot(2, 2, 4)
sns.histplot(dataset['VisITedResources_Sqrt'], bins=30, kde=True)
plt.title('Square Root Transformed Visited Resources')

plt.tight_layout()
plt.show()

# 2. Normalize or Standardize Features
# Standardization example
scaler = StandardScaler()
dataset[['raisedhands', 'VisITedResources']] = scaler.fit_transform(dataset[['raisedhands', 'VisITedResources']])

# Normalization example
normalizer = MinMaxScaler()
dataset[['raisedhands', 'VisITedResources']] = normalizer.fit_transform(dataset[['raisedhands', 'VisITedResources']])

# 3. Encode Categorical Variables
# Encoding 'Class' if it's categorical
if dataset['Class'].dtype == 'object':
    label_encoder = LabelEncoder()
    dataset['Class'] = label_encoder.fit_transform(dataset['Class'])

# One-Hot Encoding for other categorical variables
dataset_encoded = pd.get_dummies(dataset, columns=['gender', 'StageID'])

# 4. Combine Attributes
# Example of combining attributes
dataset['TotalEngagement'] = dataset['raisedhands'] + dataset['VisITedResources'] + dataset['Discussion']

# 5. Remove Outliers
# Filter only numeric columns
numeric_columns = dataset.select_dtypes(include=[np.number])

# Removing outliers using IQR
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)
IQR = Q3 - Q1
dataset_no_outliers = dataset[~((numeric_columns < (Q1 - 1.5 * IQR)) | (numeric_columns > (Q3 + 1.5 * IQR))).any(axis=1)]

# Print summary of transformations
print("Data after transformations:")
print(dataset.head())

# Optionally visualize the transformed data
plt.figure(figsize=(12, 8))
#sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix After Transformations')
plt.show()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

cat_att = ['gender','NationalITy' ,'PlaceofBirth','StageID','GradeID','SectionID','Topic','Semester','Relation','ParentAnsweringSurvey','ParentschoolSatisfaction','StudentAbsenceDays']
num_att = ['raisedhands', 'VisITedResources', 'AnnouncementsView', 'Discussion']

num_pipe = make_pipeline(
    SimpleImputer(strategy = "median"),
    StandardScaler()
)

cat_pipe = make_pipeline(
    SimpleImputer(strategy = "most_frequent"),
    OneHotEncoder(handle_unknown="ignore")
)

preprocessing = ColumnTransformer([
    ("num", num_pipe, num_att),
    ("cat", cat_pipe, cat_att)
])
```

## Hyperparameter Tuning

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

var_thresh = VarianceThreshold(threshold=0.01)

param_grids = {
    'logistic_regression_fs': {
        'feature_selection__threshold': [0.01, 0.1, 0.2],
        'model': [LogisticRegression()],
        'model__penalty': ['l1', 'l2', 'elasticnet', 'none'],
        'model__C': [0.01, 0.1, 1, 10, 100]
    },

    'random_forest_fs': {
        'feature_selection__threshold': [0.01, 0.1, 0.2],
        'model': [RandomForestClassifier()],
        'model__n_estimators': [10, 20, 25, 50, 100, 200],
        'model__max_depth': [None, 1, 2, 5, 10, 20, 30],
        'model__criterion': ['gini', 'entropy']
    },

    'svm_fs': {
        'feature_selection__threshold': [0.01, 0.1, 0.2],
        'model': [SVC()],
        'model__C': [0.01, 0.1, 1, 10, 100],
        'model__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
        'model__gamma': ['scale', 'auto', 0.001, 0.01, 0.1]
    }
}

best_models_fs = {}
best_scores_fs = {}
std_scores_fs = {}

for model_name, param_grid in param_grids.items():
    print(f"Tuning {model_name}...")


    pipeline = Pipeline([

        ('preprocessing', preprocessing),
        ('feature_selection', var_thresh),
        ('model', param_grid['model'][0])
    ])


    grid_search = GridSearchCV(pipeline, param_grid, scoring='accuracy', cv=10, n_jobs=-1)
    grid_search.fit(X, y)


    best_models_fs[model_name] = grid_search.best_estimator_
    best_scores_fs[model_name] = grid_search.best_score_

    print(f"Best Parameters for {model_name}: {grid_search.best_params_}")
    print(f"Best Score for {model_name}: {grid_search.best_score_}\n")

    # Calculate and store the standard deviation of the best model's score
    std_score = grid_search.cv_results_['std_test_score'][grid_search.best_index_]
    std_scores_fs[model_name] = std_score

    print(f"Standard Deviation for {model_name}: {std_score}\n")


results_df_fs = pd.DataFrame(list(best_scores_fs.items()), columns=['Model', 'Best Score'])
print(results_df_fs)
```
